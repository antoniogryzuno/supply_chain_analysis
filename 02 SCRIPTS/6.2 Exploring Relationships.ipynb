Since we have many qualitative variables, we need to highlight only quantitative ones

```python
quantitative_cols = [
    'Days for shipping (real)',
    'Days for shipment (scheduled)',
    'Benefit per order',
    'Sales per customer',
    'Order Item Product Price',
    'Order Item Profit Ratio',
    'Order Item Quantity',
    'Sales',
    'Order Item Total',
    'Order Profit Per Order',
    'Product Price'
]
```

```python
quantitative_data = asia[quantitative_cols]
```

```python
correlation_matrix = quantitative_data.corr()
```

```python
quantitative_data.corr()
```

### CORRELATION HEATMAP IN MATPLOTLIB

```python
# Create a correlation heatmap using matplotlib

plt.matshow(quantitative_data.corr())
plt.show()
```

```python
# Save figure
plt.matshow(quantitative_data.corr())
plt.savefig("correlation heatmap.png") 
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(quantitative_data.corr(), fignum=f.number) # type of plot
plt.xticks(range(quantitative_data.shape[1]), quantitative_data.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(quantitative_data.shape[1]), quantitative_data.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

```python
# save correlation matrix 
plt.savefig('correlation_matrix1.png', dpi=300, bbox_inches='tight')
```

### CORRELATION HEATMAP IN SEABORN

```python
# Plotting the correlation matrix using seaborn heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=1, linecolor='black')
plt.title('Correlation Matrix of Quantitative Variables', fontsize=16)
plt.show()
```

### SCATTERPLOTS

```python
# Create a scatterplot for the "Product Price" and "Sales" columns in seaborn

sns.lmplot(x = 'Product Price', y = 'Sales', data = quantitative_data)
```

We may indeed observe a very strong relationship between product price and sales. The trend line is very steep. Only one outlier can be seen. All dots are located very close to the trend line. In fact it means that increased product price leads to higher product sales. 

```python
# Create a scatterplot for the "Days for shipping (real)" and "Days for shipment (scheduled)" columns in seaborn

sns.lmplot(x = 'Days for shipment (scheduled)', y = 'Days for shipping (real)', data = quantitative_data)
```

We can see, that correlation is not strong. It has many outliers and distances between the dots. It means that the shipping schedule is not always executed in time.

'Benefit per order' <=> 'Order Item Profit Ratio' => 0.818974

```python
# Create a scatterplot for the "Benefit per order" and "Order Item Profit Ratio" columns in seaborn

sns.lmplot(x = 'Benefit per order', y = 'Order Item Profit Ratio', data = quantitative_data)
```

We may observe a strong correlation between these two variables with not so many outliers. It means that products with higher profits do bring more benefits.

### CATEGORICAL PLOTS

Creating a categorical variable that splits the "Product price" column into categories

```python
sns.histplot(quantitative_data['Product Price'], bins = 20, kde = True)
```

We can create three categories of Product Price: 1) Low Price (<200) 2) Middle Price (between 200 and 400) and High Price >400

```python
quantitative_data.loc[quantitative_data['Product Price'] < 200, 'Price category'] = 'Low price'
```

```python
quantitative_data.loc[(quantitative_data['Product Price'] >= 200) & (quantitative_data['Product Price'] < 400), 'Price category'] = 'Middle price'
```

```python
quantitative_data.loc[quantitative_data['Product Price'] >= 400, 'Price category'] = 'High price'
```

```python
quantitative_data['Price category'].value_counts(dropna = False)
```

```python
quantitative_data.columns
```

```python
# Create a categorical plot in seaborn by using categories above
sns.set(style="ticks")
sns.catplot(x="Sales", y="Product Price", hue="Price category", data=quantitative_data)
```

Above it is possible to see a catplot with prices allocated around the increasing number of sales. We can see that High Price category is located on the top and goes up. It is correlated with increased number of sales, indicated on the x-axis.

```python
#saving categorical plot
plt.savefig('categorical_plot.png', dpi=300)
```
